```python
import pandas as pd
pd.options.display.float_format = "{:,.2f}".format  #displays 2 decimal places 
df=pd.read_csv('/home/t/Downloads/A_Manuka_Perth/T4/t4/Tunnel 4_2_data.csv')
df = df.rename(columns={"Datastream Name": "stream", "Time": "time", "Value":"value"})
df.set_index("time")
```

```python
enter_df = df.loc[df['stream'] == 'BeesEntering']
enter_df = enter_df.drop('stream', 1)
enter_df = enter_df.rename(columns={"value":"bees_enter"})
print (enter_df)
enter_df.to_csv('t4_bees_enter.csv', index=False)
```

```python
exit_df = df.loc[df['stream'] == 'BeesExiting']
exit_df = exit_df.drop('stream', 1)
exit_df = exit_df.rename(columns={"value":"bees_exit"})
print (exit_df)
exit_df.to_csv('t4_bees_exit.csv', index=False)
```

```python
temp_df = df.loc[df['stream'] == 'Temperature']
temp_df = temp_df.drop('stream', 1)
temp_df = temp_df.rename(columns={"value":"temperature"})
print (temp_df)
temp_df.to_csv('t4_bees_temp.csv', index=False)
```

```python
weight_df = df.loc[df['stream'] == 'Weight']
weight_df = weight_df.drop('stream', 1)
weight_df = weight_df.rename(columns={"value":"weight"})
print (weight_df)
weight_df.to_csv('t4_weight_t4.csv', index=False)
```

```python
hum_df = df.loc[df['stream'] == 'Humidity']
hum_df = hum_df.drop('stream', 1)
hum_df = hum_df.rename(columns={"value":"humidity"})
print (hum_df)
hum_df.to_csv('t4_bees_hum.csv', index=False)
```

```python
all1_df = pd.merge(weight_df, temp_df, how='inner')
all1_df
```

```python
all2_df = pd.merge(all1_df, hum_df, how='inner')
all3_df = pd.merge(all2_df, enter_df, how='inner')
all4_df = pd.merge(all3_df, exit_df, how='inner')
all4_df.to_csv('all_tunn1.csv', index=False)
all4_df.head(50)
```

```python
#reverse data - upside down
all5_df=all4_df.iloc[::-1]
all5_df
```

```python
#reset index 
all5_df = all5_df.reset_index(drop=True)
all5_df
# write to csv REMEMBER TO CHANGE !!!!!! THE NAME FOR CSV !!!!!!! 

all5_df.to_csv('tunnel4_new2.csv', index=False)
```
